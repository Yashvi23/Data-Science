```
#To get column name
df = pd.DataFrame(housing_data.data)
df.columns = housing_data.feature_names
df
#Below mentioned data just contains i/p one row needs to be added for o/p
```

```
df['Price'] = housing_data.target
df.head() #only first 5 rows
```

```
#Method1
x = df.iloc[:,0:13].values#it will go from 0 to 12
y=df.iloc[:,13].values
print(x)
print(y)
```

```
# Data in the form i/p or o/p
#Splitting of Data = Training+Testing
#4 values:
#input Training
#output Training
#input Training
#output Training
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state = 50) #dont change the order since first o/p is train second is test then 3rd is the output of y
#i.e, train and test # test value will be 30% and train value will be 70% if random state is 1, then seed value is 1 and it will take alternatevalues
#for train and test and if it is 2 then every 3rd data will be chosen
```

```
x_train.shape
```

```
x_test.shape
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train) #only train data
y_pred = model.predict(x_test)
y_pred #predicted output
```

```
y_test #real o/p of testing values
```

```
#To check the accuracy of theo/p obtained
from sklearn import metrics

#mean squared error- error between the real o/p and predicted o/p
metrics.mean_squared_error(y_test,y_pred)
```

```
#Accuracy Score
metrics.r2_score(y_test,y_pred)#best score is 1
```
