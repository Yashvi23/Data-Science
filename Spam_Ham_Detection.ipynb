```
x=df.iloc[:,1].values #we will be converting it to vector#Message column as i/p
y=df.iloc[:,0].values#label o/p
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```
print(x_train.shape)
print(x_test.shape)
```

```
#Steps in NlP ML
#1. Gather the data(Create a dataframe)
#2. Divide intoi/p and o/p
#3. split into train and test
#4. Convert i/p into numericals- Count Vectorizer+TFIDF Transformer or TFIDF Vectorizer

#TFIDF Vectorizer - Count Vectorizer + TFIDF Transformer
#5. Calling the algorithm and fitting the model
#6. Evaluating the model
```

```
#Pipeline=  4th+5th step
```

```
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB #o/p in the form of probability
#Pipeline([('Variable 1,'Method 1()),('Variable 2','Method 2()')]) #A function with a list consisting of tuple
```

```
text_model = Pipeline([('tfidf',TfidfVectorizer()),('model',MultinomialNB())])
```

```
text_model.fit(x_train,y_train)
```

```
y_pred = text_model.predict(x_test)
y_pred
```

```
y_test
```

```
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```
accuracy_score(y_pred,y_test)*100
```

```
confusion_matrix(y_pred,y_test)
```

```
print(classification_report(y_pred,y_test))
```
