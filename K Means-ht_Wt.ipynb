```
import matplotlib.pyplot as plt
plt.scatter(df['Height'],df['Weight'])#HT=Xaxis,Wt=Yaxis
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```
x = df.iloc[:,[0,1]].values
x
```

```
#Elbow Method K=2
#Its an assumption

```

```
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2,random_state=0)
model.fit(x)
y_pred=model.predict(x)
y_pred
```

```
#pLOTTING THE CLUSTER
#1. CLUSTER-0-10 VALUES(10 FOR HT,x AND 10 FOR WT,y)
#HEIGHT VALUES(X)
x[y_pred == 0,0] #values of cluster 0 of 0th column
```

```
#Weight VALUES(X)
x[y_pred == 0,1] #values of cluster 0 of 1th column
```

```
#HEIGHT VALUES(X)
x[y_pred == 1,0] #values of cluster 1 of 0th column
```

```
#Weight VALUES(X)
x[y_pred == 1,1] #values of cluster 1 of 1th column
```

```
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1],c='r',label='Cluster 0')
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1],c='y',label='Cluster 1')
plt.legend()
plt.show()
```

```
#Elbow Method
#Determining the value of k
sse = []#Sum of Squared errors. A list to store the value
for i in range(1,13): #Take a loop from 1 to maz value
 model = KMeans(n_clusters =i)
 model.fit(x)
 sse.append(model.inertia_)
sse
```

```
plt.plot(range(1,13),sse)
plt.xlabel('No of Cluster')
plt.ylabel('SSE')
plt.plot
```
